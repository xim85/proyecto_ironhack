```python
import pandas as pd
import numpy as np
```

### Cargar DFs

```python
df_temperatura = pd.read_csv('df_temperatura.csv')
df_hoteles = pd.read_csv('df_hoteles.csv')
df_turismo_interno = pd.read_csv('df_turismo_interno.csv')
df_turismo_residentes = pd.read_csv('df_turismo_residentes.csv')
df_gasto_medio = pd.read_csv('df_turismo_residentes.csv')
```

### Mapeo Provincias

```python
df_hoteles['Provincias'].unique()
```

```python
df_temperatura['PROVINCIA'].unique()
```

```python
df_turismo_residentes
```

```python
mapeo_provincias = {
    'Palmas, Las': 'Las Palmas',
    'Coruña, A': 'A Coruña',
    'Alicante/Alacant': 'Alicante',
    'Castellón/Castelló': 'Castellón',
    'Valencia/València': 'Valencia',
    'Araba/Álava' : 'Álava',
    'Comunidad de Madrid' : 'Madrid',
    'Región de Murcia' : 'Murcia',
    'Comunidad Foral de Navarra': 'Navarra',
    'Principado de Asturias' : 'Asturias'
}
df_hoteles['Provincias'] = df_hoteles['Provincias'].replace(mapeo_provincias)
df_temperatura['PROVINCIA'] = df_temperatura['PROVINCIA'].replace(mapeo_provincias)
df_turismo_interno['PROVINCIA_DESTINO'] = df_turismo_interno['PROVINCIA_DESTINO'].replace(mapeo_provincias)
df_turismo_interno['PROVINCIA_ORIGEN'] = df_turismo_interno['PROVINCIA_ORIGEN'].replace(mapeo_provincias)
```

### Unir df_turismo_interno con df_hoteles

```python
df_unido = pd.merge(
    df_turismo_interno,
    df_hoteles,
    left_on=['AÑO', 'MES', 'CCAA_DESTINO', 'PROVINCIA_DESTINO'],
    right_on=['Año', 'Mes', 'Comunidades y Ciudades Autónomas', 'Provincias'],
    how='inner'
)

# 2. Eliminar las columnas duplicadas después de la unión
df_unido = df_unido.drop(columns=['Año', 'Mes', 'Comunidades y Ciudades Autónomas', 'Provincias'])
```

```python
df_unido.shape
```

Renombrar columnas

```python
df_unido.rename(columns={'AÑO': 'Año','MES': 'Mes', 
    'CCAA_ORIGEN': 'CCAA_Origen', 
    'CCAA_DESTINO' : 'CCAA_Destino', 
    'PROVINCIA_ORIGEN' : 'Provincia_Origen',
    'PROVINCIA_DESTINO' : 'Provincia_Destino', 
    'TURISTAS' : 'Turistas', 
    'PERNOCTACIONES' : 'Pernoctaciones',
    'ESTANCIA_MEDIA' : 'Estancia_Media'}, inplace=True)
df_unido.columns
```

### Unir con df_temperatura

```python
df_temperatura.columns
```

```python
df_unido.columns
```

```python
df_final = pd.merge(
    df_unido,
    df_temperatura,
    left_on=['Año', 'Mes', 'CCAA_Destino', 'Provincia_Destino'],
    right_on=['AÑO', 'MES', 'CCAA', 'PROVINCIA'],
    how='inner'
)
df_final = df_final.drop(columns=['AÑO', 'MES', 'PROVINCIA', 'CCAA'])
```

```python
df_final.shape
```

```python
df_final
```

### Unir con gasto medio

```python
df_gasto_medio
```

```python
df_final = df_final.drop_duplicates()
```

```python
df_final.head(60)
```

```python
df_final = pd.merge(
    df_final,
    df_gasto_medio[['Destino','Año','Mes','Gasto medio diario por persona']],
    left_on=['CCAA_Destino', 'Año', 'Mes'],
    right_on=['Destino', 'Año', 'Mes'],
    how='left'
)
df_final.drop(columns=['Destino_x','Destino_y', 'Destino'], inplace=True, errors='ignore')
```

```python
df_final.isna().sum()
```

```python
df_final.dropna(subset=['Gasto medio diario por persona'], inplace=True)
```

```python
df_final.to_csv('df_final.csv', index= False)
```
